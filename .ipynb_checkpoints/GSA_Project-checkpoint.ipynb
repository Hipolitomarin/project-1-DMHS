```python
import pandas as pd
import re

def clean_time(val):
    """
    Standardize time formats like '11h00' â†’ '11:00', keep words like 'Afternoon'.
    Handles NaN safely.
    """
    if pd.isna(val):
        return val  # Keep NaN as is

    val = str(val).strip()  # Convert to string, remove surrounding spaces
    
    # Regex to detect time patterns like 11h00, 16h30, etc.
    time_pattern = r'^(\d{1,2})h(\d{2})$'
    match = re.match(time_pattern, val, re.IGNORECASE)
    
    if match:
        hour = match.group(1).zfill(2)
        minute = match.group(2)
        return f"{hour}:{minute}"
    
    return val  # Return other values unchanged

# Example application
df['Cleaned_Time'] = df['Time'].apply(clean_time)
```

```python
def clean_time(val):
    if pd.isna(val):
        return val

    val = str(val).strip().lower()

    # 1. Pattern: 4-digit time like '1200hrs', '1100hr', '0945hrs'
    match = re.match(r'^(\d{2})(\d{2})h?r?s?$', val)
    if match:
        hour = match.group(1).zfill(2)
        minute = match.group(2)
        return f"{hour}:{minute}"

    # 2. Pattern: Time with decimal separator e.g. '16.30hrs' or '12.11 hrs'
    match = re.match(r'^(\d{1,2})[.:](\d{2})h?r?s?$', val)
    if match:
        hour = match.group(1).zfill(2)
        minute = match.group(2)
        return f"{hour}:{minute}"

    # 3. Pattern: '11hr15' or '16hr15'
    match = re.match(r'^(\d{1,2})hr(\d{2})$', val)
    if match:
        hour = match.group(1).zfill(2)
        minute = match.group(2)
        return f"{hour}:{minute}"

    # 4. Pattern: Full hour only e.g. '1100hr'
    match = re.match(r'^(\d{2})(\d{2})hr$', val)
    if match:
        hour = match.group(1).zfill(2)
        minute = match.group(2)
        return f"{hour}:{minute}"

    # 5. Known irrelevant or unclear values â†’ Replace with NaN
    if val in ['?', 'unknown', 'not stated', 'am', 'pm', 'after 1200hr']:
        return None

    # 6. Leave descriptive labels like 'Afternoon', 'Morning', etc.
    return val
```

```python
df['Cleaned_Time'] = df['Cleaned_Time'].apply(clean_time)
```

```python
import re

def clean_time(value):
    if pd.isna(value):
        return None

    value = str(value).strip().lower()

    # Recognize formats like 1200hrs, 16h15, 11.30hr, 15.5 etc.
    time_match = re.match(r'(\d{1,2})[h:.]?(\d{2})?', value)

    if time_match:
        hour = time_match.group(1)
        minute = time_match.group(2) if time_match.group(2) else "00"
        return f"{hour.zfill(2)}:{minute.zfill(2)}"

    # Keep descriptive times like Morning, Afternoon, etc.
    if any(phrase in value for phrase in ['morning', 'afternoon', 'evening', 'night', 'dusk', 'midday', 'early', 'late']):
        return value.title()

    return None
```

```python
df['Cleaned_Time'] = df['Time'].apply(clean_time)
```
